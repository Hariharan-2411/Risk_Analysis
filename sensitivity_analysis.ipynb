```python
import matplotlib.pyplot as plt
import numpy as np

# Define symbolic EV equations
def ev_predict_real(p):
    return 259 * p + 49

def ev_predict_fake(p):
    return -64 * p + 70

# Generate p values for plotting
p_vals = np.linspace(0, 1, 100)
ev_real_vals = [ev_predict_real(p) for p in p_vals]
ev_fake_vals = [ev_predict_fake(p) for p in p_vals]

# Intersection point from symbolic calculation
p_intersection = 21 / 323
ev_intersection = ev_predict_real(p_intersection)

# Plotting the graph
plt.figure(figsize=(10, 6))
plt.plot(p_vals, ev_real_vals, label='EV(Predict Real) = 259p + 49', linewidth=2)
plt.plot(p_vals, ev_fake_vals, label='EV(Predict Fake) = -64p + 70', linewidth=2)
plt.axvline(p_intersection, linestyle='--', color='gray', label=f'Intersection p ≈ {p_intersection:.3f}')
plt.axhline(ev_intersection, linestyle='--', color='gray')
plt.title('Sensitivity Analysis')
plt.xlabel('p = Probability News is Real')
plt.ylabel('Expected Value (Payoff)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
